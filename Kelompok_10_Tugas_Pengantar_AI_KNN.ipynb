```python
import numpy as np
import pandas as pd
```

```python
# Membaca File mobil.xls 
df = pd.read_excel('https://github.com/mzakibhr/kNN-PAI-Tupro3/blob/main/mobil.xls?raw=true')
df.head(17)
```

```python
# Data Processing

# Mengganti nama kolom untuk mempermudah pembuatan program
data = df.rename({'Harga (Ratus Juta)': 'Harga'}, axis=1)

# Melakukan Drop Kolom "Nama Mobil"
data = data.drop('Nama Mobil', axis=1) 

# Normalisasi data "Harga"
for kolom in range(len(data)):
    data['Harga'][kolom] = (data['Harga'][kolom] - (data['Harga']).min()) / ((data['Harga']).max() - (data['Harga']).min())

# Definisi rumus metode perhitungan jarak
def jarak_euclidean(x, y): # Perhitungan Jarak Euclidean
    jarak = 0
    for i in range(len(y)):
        jarak += (x[i] - y[i])**2

    return jarak ** (1/2)

def jarak_manhattan(x, y): # Perhitungan Jarak Manhattan
    jarak = 0
    for i in range(len(x)):
        jarak += abs(x[i] - y[i])

    return jarak
 
def jarak_minkowski(x, y, p=2): # Perhitungan Jarak Minkowski
    jarak = 0
    for i in range(len(x)):
        jarak += abs(x[i] - y[i]) ** p

    return jarak ** (1/p)

def jarak_supremum(x, y): # Perhitungan Jarak Supremum
    jarak = []
    for i in range(len(x)):
        jarak.append(abs(x[i] - y[i]))

    return max(jarak)

# Klasifikasi K-Nearest Neighbour
def jarak(x, y, k):
    euclidean, manhattan, minkowski, supremum = [], [], [], []

    for i in range(len(x)):
        minkowski.append([jarak_minkowski(x.iloc[i].tolist(), y), df['Nama Mobil'][i]])
        manhattan.append([jarak_manhattan(x.iloc[i].tolist(), y), df['Nama Mobil'][i]])
        euclidean.append([jarak_euclidean(x.iloc[i].tolist(), y), df['Nama Mobil'][i]])
        supremum.append([jarak_supremum(x.iloc[i].tolist(), y), df['Nama Mobil'][i]])

    # Sort List
    euclidean = sorted(euclidean)[:k]
    manhattan = sorted(manhattan)[:k]
    minkowski = sorted(minkowski)[:k]
    supremum = sorted(supremum)[:k]

    return euclidean, manhattan, minkowski, supremum

def knn(x, y, k=3): # Inisiasi Variabel k = 3, Untuk Memilih 3 Mobil Terbaik
    euclidean, manhattan, minkowski, supremum = jarak(x, y, k)
    hasil = [['Euclidean', euclidean], ['Manhattan', manhattan], ['Minkowski', minkowski], ['Supremum', supremum]]

    return hasil

# Main Program
ukuran = int(input("Masukkan Ukuran     : "))
kenyamanan = int(input("Masukkan Kenyamanan : "))
irit = int(input("Masukkan Irit       : "))
kecepatan = float(input("Masukkan Kecepatan  : "))
harga = float(input("Harga               : "))

hasil = knn(data, [ukuran, kenyamanan, irit, kecepatan, harga])

for metode in hasil:
      print("\n====================================================")
      print(f'Metode: {metode[0]}')
      print("====================================================")
    
      mobil = []
      for i, solusi in enumerate(metode[1]):
          print(f'{i+1}. {solusi[1]}, solusi: {round(solusi[0], 2)}')
          mobil.append(solusi[1])

mobil = pd.DataFrame(mobil)
mobil.to_excel(f'rekomendasi.xls', index=False, header=False)
```
